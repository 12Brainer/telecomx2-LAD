```python
# importar liberias necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# cargar los datos
df = pd.read_csv('/content/datos_tratados.csv')
df.head()
```

```python
# ver las columnas
df.columns
```

```python
# Informaci√≥n general
df.info()
```

##üõ†Ô∏è Pre procesamiento

‚úîÔ∏è Remover columnas irrelevantes

```python
# Eliminar columna que no se usar√°
df = df.drop(columns=['CustomerID'])
```

```python
# Mostrar el DataFrame
df
```

```python
# Unificar valores de servicios
cols_to_fix = [
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

for col in cols_to_fix:
    df[col] = df[col].replace({'No internet service': 'No'})
```

‚úîÔ∏è Verificaci√≥n nuevamente de los valores nulos

```python
df.isnull().sum()
```

## üìä EDA

‚úîÔ∏è Correlaci√≥n entre variables numericas

```python
# Seleccionar variables num√©ricas
df_numeric = df.select_dtypes(include=["int64", "float64"])

# Convertir Churn a binario
df_numeric["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})

# Matriz de correlaci√≥n
corr_matrix = df_numeric.corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlaci√≥n - Variables Num√©ricas")
plt.show()
```

```python
# Eliminar columnas
df = df.drop(columns=["ChargesDaily", "ChargesTotal"])
```

2Ô∏è‚É£ Analisis de churn por categoria (variables categoricas)

üîπ Seleccionar variables categoricas

```python
# Obtener columnas categ√≥ricas (excepto Churn)
categorical_cols = df.select_dtypes(include=["object"]).columns
categorical_cols = categorical_cols.drop("Churn")
```

üîπ Tabla de proporci√≥n por categoria

```python
# An√°lisis porcentual por categor√≠a vs Churn
for col in categorical_cols:
    print(f"\nüìå An√°lisis de variable: {col}")
    display(pd.crosstab(df[col], df["Churn"], normalize="index") * 100)
```

An√°lisis Dirigido

üéØ Metodo recomendado: Qui-cuadrado (Chi-Square Test)

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import chi2

# Copia del DataFrame
df_rank = df.copy()

# Churn a binario
df_rank["Churn"] = df_rank["Churn"].map({"Yes": 1, "No": 0})

# Columnas categ√≥ricas
categorical_cols = df_rank.select_dtypes(include=["object"]).columns

# Label Encoding
le = LabelEncoder()
for col in categorical_cols:
    df_rank[col] = le.fit_transform(df_rank[col])

# Variables y objetivo
X_cat = df_rank[categorical_cols]
y = df_rank["Churn"]

# Prueba Chi-cuadrado
chi_scores, p_values = chi2(X_cat, y)

# Ranking de importancia
chi_df = pd.DataFrame({
    "Variable": categorical_cols,
    "Chi2 Score": chi_scores,
    "p-value": p_values
}).sort_values(by="Chi2 Score", ascending=False)

chi_df
```

```python
# Variables significativas (p < 0.05)
significant_vars = chi_df[chi_df["p-value"] < 0.05]

print("\n‚úÖ Variables significativas:")
display(significant_vars)
```

```python
# Eliminar columnas
df = df.drop(columns=["Gender", "PhoneService"])
```

## üéØ Split

```python
# Separar variables y objetivo
X = df.drop("Churn", axis=1)
y = df["Churn"]
```

2.Split

```python
from sklearn.model_selection import train_test_split

# Divisi√≥n train / test
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

3.One-Hot Encoding




```python
# Aplicar One-Hot Encoding en treno
X_train = pd.get_dummies(X_train, drop_first=True)

# Aplicar en teste
X_test = pd.get_dummies(X_test, drop_first=True)

# Garantir que treno y teste tengan las mismas columnas
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)

X_train.head()
```

```python
# Informaci√≥n del set de entrenamiento
X_train.info()
```

4. Verificar desbalanceamiento

```python
# Distribuci√≥n porcentual de la variable objetivo
y_train.value_counts(normalize=True) * 100
```

##Entrenamiento

Modelo 1 ‚Äî Regresion logistica

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Entrenar modelo
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)

# Predicci√≥n
y_pred_log_normal = log_model.predict(X_test)

# Evaluaci√≥n
print("Regresi√≥n Log√≠stica")
print(classification_report(y_test, y_pred_log_normal))
```

An√°lisis de regresi√≥n logistica


El modelo de regresi√≥n log√≠stica obtiene una exactitud del 80%, mostrando un buen desempe√±o general. Predice muy bien a los clientes que no cancelan (recall del 90%), pero tiene dificultades para identificar a los clientes que s√≠ cancelan, ya que solo detecta el 52% de los casos de churn. Esto indica que, aunque el modelo funciona bien como punto de partida, no es √≥ptimo para detectar la cancelaci√≥n, que suele ser el objetivo principal del an√°lisis, por lo que ser√≠a recomendable aplicar t√©cnicas para mejorar el recall de la clase Yes.

Modelo 2 ‚Äî Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

# Crear y entrenar el modelo Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Realizar predicciones
y_pred_rf = rf_model.predict(X_test)

# Evaluar el modelo
print("Random Forest")
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.linear_model import LogisticRegression

# Modelo con balanceo de clases
log_model = LogisticRegression(max_iter=1000, class_weight="balanced")
log_model.fit(X_train, y_train)

# Predicci√≥n
y_pred_log = log_model.predict(X_test)

# Evaluaci√≥n
print(classification_report(y_test, y_pred_log))
```

An√°lisis de Random Forest y balanceo de clases

El Random Forest muestra una accuracy m√°s alta (78%), con muy buen desempe√±o para la clase No (recall 90%), pero presenta una baja capacidad para detectar churn, ya que solo identifica el 48% de los clientes que s√≠ cancelan. Esto indica que el modelo est√° sesgado hacia la clase mayoritaria y pierde muchos casos importantes de cancelaci√≥n.

Por otro lado, la Regresi√≥n Log√≠stica con class_weight="balanced" reduce la accuracy total (74%), pero mejora significativamente la detecci√≥n de churn, alcanzando un recall del 79% en la clase Yes, lo que significa que identifica correctamente a la mayor√≠a de los clientes que cancelan. Aunque comete m√°s falsos positivos, este modelo es m√°s adecuado cuando el objetivo principal es prevenir churn, ya que prioriza no dejar pasar clientes en riesgo.

Informe detallado

```python
# 1. Preparar los datos extra√≠dos de tus reportes
data = {
    'Modelo': ['LogReg Normal', 'Random Forest', 'LogReg Balanceada'],
    'Precision': [0.65, 0.62, 0.51],
    'Recall': [0.52, 0.48, 0.79],
    'F1-Score': [0.58, 0.54, 0.62]
}

df = pd.DataFrame(data)

# 2. Configurar el gr√°fico
plt.style.use('seaborn-v0_8-darkgrid') # Estilo visual agradable
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.25
x = np.arange(len(df['Modelo']))

# 3. Crear las barras
bars1 = ax.bar(x - bar_width, df['Precision'], bar_width, label='Precision', color='#4C72B0')
bars2 = ax.bar(x, df['Recall'], bar_width, label='Recall', color='#55A868')
bars3 = ax.bar(x + bar_width, df['F1-Score'], bar_width, label='F1-Score', color='#C44E52')

# 4. Personalizar textos y ejes
ax.set_ylabel('Puntuaci√≥n (0 a 1)', fontweight='bold', fontsize=12)
ax.set_title('Comparaci√≥n de M√©tricas para la Clase "Yes" (Minoritaria)', fontweight='bold', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(df['Modelo'], fontsize=11)
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# 5. Funci√≥n para a√±adir los n√∫meros sobre cada barra
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # Desplazamiento vertical
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10)

add_labels(bars1)
add_labels(bars2)
add_labels(bars3)

# 6. Mostrar el gr√°fico
plt.ylim(0, 1.05)
plt.tight_layout()
plt.show()
```

Se eligi√≥ el modelo con class_weight= balanced para compensar el desbalance de los datos. Aunque el accuracy general baja ligeramente, este ajuste permite que el modelo deje de ignorar la clase minoritaria y logre detectar el 79% de los casos 'Yes' Recall, superando por mucho el desempe√±o de los modelos est√°ndar que apenas alcanzaban el 52%.
